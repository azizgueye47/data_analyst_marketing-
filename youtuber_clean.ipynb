```python
import pandas as pd
from sqlalchemy import create_engine
youtuber =pd.read_csv('https://raw.githubusercontent.com/sdw-online/top_uk_youtubers_2024/main/assets/datasets/youtube_data_from_python.csv'
                      ,encoding='utf-8')
```

```python
youtuber.head()
```

```python
youtuber.info()
```

```python
youtuber=youtuber[['NOMBRE','total_subscribers','total_views','total_videos']].copy()
```

```python
channel_name =[]
for compte in youtuber.NOMBRE :
    channel_name.append(compte[:compte.find("@")])
youtuber['channel_name']=channel_name
youtuber = youtuber.drop(columns=['NOMBRE'])        
```

```python
youtuber
```

```python
engine = create_engine('postgresql://postgres:@@@@@2000@localhost:5432/postgres')
youtuber.to_sql('youtuber_table', engine, if_exists='replace', index=False)
```
